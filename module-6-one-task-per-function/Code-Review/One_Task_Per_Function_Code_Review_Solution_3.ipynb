## Problem:

Given the following data definitions, design a function that takes a list of students and a program and returns a list of students that are in the given program.

You must also design any helper functions needed according to the helper rules. 

```python
# This cell is uneditable so you can refer back to the original version of the solution.

from enum import Enum
from typing import NamedTuple, List, Optional
from cs103 import *

Program = str
# interp. a program at UBC
P1 = "BMUS"
P2 = "BKIN"
P3 = "BSC"
P4 = "BMS"

@typecheck
def fn_for_program(p: Program) -> ...: # template based on simple atomic
    return ...(p)

    

CourseRecord = NamedTuple('CourseRecord', [('dept', str),    # four characters
                                           ('code', int),    # in range [100, 999]
                                           ('grade', Optional[int])]) # in range [0, 100]
# interp. a course record for a student at UBC with their final grade ('grade') where None 
#         means the course is not yet complete, and the course's department ('dept') and course 
#         code ('code'). 
CR1 = CourseRecord('CPSC', 103, 99)
CR2 = CourseRecord('ENGL', 112, 84)

@typecheck
def fn_for_course_record(cr: CourseRecord) -> ...: # template based on compound
    return ...(cr.dept,
               cr.code,
               cr.grade)

# List[CourseRecord]
# interp. a list of course records

L0 = []
L1 = [CR1, CR2]

@typecheck
def fn_for_locr(locr: List[CourseRecord]) -> ...: # template based on arbitrary-sized and the
    # description of the acc                      # reference rule
    acc = ... # type: ...
    for cr in locr:
        acc = ...(acc, fn_for_course_record(cr))
    return ...(acc)


Student = NamedTuple ('Student', [('name', str), 
                                  ('stu_num', int),  # in range [11111111, 99999999] 
                                  ('course_records', List[CourseRecord]), 
                                  ('program', Program)])
# interp. a student with their name, student number, course records, and program

S1 = Student("Jessica Jones", 25252325, [], "BSC")
S2 = Student("Cindy Lu", 87774488, [CR1, CR2], "BMUS")
S3 = Student("Raven Quan", 99996699, [CR1], "BSC")

@typecheck
def fn_for_student(s: Student) -> ...: # template based on compound and reference rule (two times)
    return ...(s.name,
               s.stu_num,
               fn_for_locr(s.course_records),
               fn_for_program(s.program))


# List[Student]
# interp. a list of students

LOS1 = []
LOS2 = [S1, S2]
LOS3 = [S1, S2, S3]

@typecheck
def fn_for_los(los: List[Student]) -> ...: # template based on arbitrary-sized and the reference rule
    # description of the acc
    acc = ... # type: ...
    for s in los:
        acc = ...(acc, fn_for_student(s))
    return ...(acc)
```

```python
# This cell is uneditable so you can refer back to the original version of the solution.

#solution

@typecheck
def equal_programs(p1: Program, p2: Program) -> bool:
    """
    return True if p1 is equal to p2
    """
    # return False #stub
    # template from Program
    if p1 == p2:
        return True 
    else: 
        return False

@typecheck
def in_program(s: Student, p: Program) -> bool:
    """
    return True if the student is in program p
    """
    # return False # stub
    # template from Student with added atomic parameter
    
    return equal_programs(s.program, p)

@typecheck
def students_in_same_program (los: List[Student], p: Program) -> List[Student]:
    """
    return a list containing all students from los that are in the same program as p
    """
    # return [] #stub
    # template from List[Student] with additional parameter
    
    # result contains the list of student in the program so far
    result = [] # type: List[Student]
    for s in los:
         if (in_program(s, p)):
                result.append(s)
    return result

# Start testing
start_testing()

S1 = Student("Jessica Jones", 25252325, [], "BSC")
S2 = Student("Cindy Lu", 87774488, [CR1, CR2], "BMUS")
S3 = Student("Raven Quan", 99996699, [CR1], "BSC")
S4 = Student("John Smith", 53383838, [CR2], "BA") 

LOS1 = []
LOS2 = [S1, S2]
LOS3 = [S1, S2, S3]
LOS4 = [S1, S2, S3, S4]

# Tests for equal_programs
expect(equal_programs("", ""), True)
expect(equal_programs("BMUS", ""), False)
expect(equal_programs("", "BMUS"), False)
expect(equal_programs("BMUS", "BKIN"), False)
expect(equal_programs("BKIN", "BSC"), False)
expect(equal_programs("BSC", "BSC"), True)

# Tests for in_program
expect(in_program(S1, "BSC"), True)
expect(in_program(S1, "BA"), False)

# Tests for students_in_same_program

expect(students_in_same_program(LOS1, "BA"), [])
expect(students_in_same_program(LOS2, "BA"), [])
expect(students_in_same_program(LOS3, "BSC"), [S1, S3])
expect(students_in_same_program(LOS4, "BA"), [S4])

# Show testing summary
summary()
```
