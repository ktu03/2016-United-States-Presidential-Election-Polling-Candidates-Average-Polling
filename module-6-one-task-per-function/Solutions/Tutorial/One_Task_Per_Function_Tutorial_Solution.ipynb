## Problem 1

Design a function that takes a city, a grant range (in the form of a minimum and a maximum grant amount), and a list of government grants and returns **all the government grants** in the list within the grant range (including the limits) awarded in that city.

In particular, the **function MUST clearly implement these three steps**:
1. First, it will find the government grants in the given city,
2. Then it will find which of those government grants received grant amounts **no more than** a maximum value, and
3. Then it will find which of those government grants received grant amounts **no less than** a minimum value.

Remember to follow the HtDF recipe and the helper rules.

```python
# your solution goes here

@typecheck
def filter_on_city_and_grant_amount(city: str, min_amt: int, max_amt: int, logg: List[GovernmentGrant]) -> List[GovernmentGrant]:
    """
    return a list of grants from logg that are from the given city and 
    have amounts in the range [min_amt, max_amt] (inclusive)
    """
    #return [] #stub
    # template based on function composition
    
    # You aren't required to include a plan in comments, 
    # and your solution can be all in one line. But..
    # we like the way this looks and reads!
    
    
    # Plan:
    # Step 1: Find the government grants from logg in the given city, call it city_grants
    # Step 2: Find which of the government grants in city_grants received grant amounts 
    #         no more than max_amt, call it small_city_grants
    # Step 3: Find which of the government grants in small_city_grants received grant amounts 
    #         no less than min_amt, call it city_grants_in_range
    # Step 4: Return city_grants_in_range

    
    # Implementation:
    # Step 1: Find the government grants from logg in the given city, call it city_grants
    city_grants = from_city_only(city, logg)
    # Step 2: Find which of the government grants in city_grants received grant amounts 
    #         no more than max_amt, call it small_city_grants
    small_city_grants = grants_no_more_than(max_amt, city_grants)
    # Step 3: Find which of the government grants in small_city_grants received grant amounts 
    #         no less than min_amt, call it city_grants_in_range
    city_grants_in_range = grants_no_less_than(min_amt, small_city_grants)
    # Step 4: Return city_grants_in_range
    return city_grants_in_range
    
    
    # The alternate one-line version would be:
    # return grants_no_less_than(min_amt, grants_no_more_than(max_amt, from_city_only(city, logg)))
    
    # Can you see how that looks like the four lines above,
    # but with each variable name substituted for the expression
    # from which we got its value?

@typecheck
def grants_no_more_than(max_amt: int, logg: List[GovernmentGrant]) -> List[GovernmentGrant]:
    """
    return a list of grants from logg that have amounts no more than max_amt (inclusive)
    """
    # return [] #stub
    # template from List[GovernmentGrant] with additional parameter
    
    # acc stores the result so far
    acc = [] # type: List[GovernmentGrant]
    
    for gg in logg:
        if amt_is_no_more_than(max_amt, gg):
            acc.append(gg)

    return acc

@typecheck
def grants_no_less_than(min_amt: int, logg: List[GovernmentGrant]) -> List[GovernmentGrant]:
    """
    return a list of grants from logg that have amounts no less than min_amt (inclusive)
    """
    # return [] #stub
    # template from List[GovernmentGrant] with additional parameter
    
    # acc stores the result so far
    acc = [] # type: List[GovernmentGrant]
    
    for gg in logg:
        if amt_is_no_less_than(min_amt, gg):
            acc.append(gg)

    return acc

@typecheck
def amt_is_no_more_than(max_amt: int, gg: GovernmentGrant) -> bool:
    """
    return True if gg has an amount less than or equal to a specified `max_amt` and False otherwise
    """
    #return False #stub
    # template from GovernmentGrant with additional parameter
    return gg.amt <= max_amt

@typecheck
def amt_is_no_less_than(min_amt: int, gg: GovernmentGrant) -> bool:
    """
    return True if gg has an amount greater than or equal to a specified `min_amt` and False otherwise
    """
    #return False #stub
    # template from GovernmentGrant with additional parameter
    return gg.amt >= min_amt

@typecheck
def from_city_only(city: str, logg: List[GovernmentGrant]) -> List[GovernmentGrant]:
    """
    return a list of grants from logg that are from a given city
    """
    # return [] #stub
    # template from List[GovernmentGrant] with additional parameter
    
    # acc stores the result so far
    acc = [] # type: List[GovernmentGrant]
    
    for gg in logg:
        # It is perhaps a judgment call whether we need
        # a helper to check if gg is from area, but in
        # CPSC 103, we ask you to create this helper!
        if from_city(city, gg):
            acc.append(gg)
    
    return acc

@typecheck
def from_city(city: str, gg: GovernmentGrant) -> bool:
    """
    return True if gg is from the given city, False otherwise
    """
    #return False #stub
    # template from GovernmentGrant with additional parameter
    return gg.city == city

start_testing()

expect(filter_on_city_and_grant_amount("Vancouver", 0, 80000, []), []) 
expect(filter_on_city_and_grant_amount("Vancouver", 0, 20000, L1), [ALL_BODIES])
expect(filter_on_city_and_grant_amount("Vancouver", 20000, 80000, L1), [AFRICAN_DESCENT_SOCIETY])
expect(filter_on_city_and_grant_amount("Abbotsford", 0, 79999, L1), [BMX, BIA, AIRSHOW])
expect(filter_on_city_and_grant_amount("Abbotsford", 0, 80000, L1), [AGRIFAIR, BMX, BIA, AIRSHOW])
expect(filter_on_city_and_grant_amount("Abbotsford", 60000, 60000, L1), [BIA, AIRSHOW])
expect(filter_on_city_and_grant_amount("Richmond", 0, 30000, L1), [CHINESE_ORCHESTRA])
expect(filter_on_city_and_grant_amount("Richmond", 0, 31500, L1), [YOUTH_MUSIC, CHINESE_ORCHESTRA])
expect(filter_on_city_and_grant_amount("Richmond", 0, 3000, L1), [])
expect(filter_on_city_and_grant_amount("Richmond", 80001, 100000, L1), [])

expect(grants_no_less_than(0, []), [])
expect(grants_no_less_than(25001, L1), [AGRIFAIR, BMX, BIA, AIRSHOW, BARRACUDAS, YOUTH_MUSIC])
expect(grants_no_less_than(0, L1), L1)

expect(grants_no_more_than(100000, []), [])
expect(grants_no_more_than(1, L1), [])
expect(grants_no_more_than(25000, L1), [ARTS_COUNCIL, AFRICAN_DESCENT_SOCIETY, ALL_BODIES, CHINESE_ORCHESTRA])
expect(grants_no_more_than(100000, L1), L1)

expect(amt_is_no_less_than(0, CHINESE_ORCHESTRA), True)
expect(amt_is_no_less_than(29499, BMX), True)
expect(amt_is_no_less_than(29500, BMX), True)
expect(amt_is_no_less_than(29501, BMX), False)
expect(amt_is_no_less_than(51000, BARRACUDAS), True)
expect(amt_is_no_less_than(52000, BARRACUDAS), False)

expect(amt_is_no_more_than(5000, CHINESE_ORCHESTRA), True)
expect(amt_is_no_more_than(4999, CHINESE_ORCHESTRA), False)
expect(amt_is_no_more_than(52000, BARRACUDAS), True)
expect(amt_is_no_more_than(51000, BARRACUDAS), False)

expect(from_city_only("Abbotsford", []), []) 
expect(from_city_only("Abbotsford", L1), [AGRIFAIR, BMX, BIA, AIRSHOW])
expect(from_city_only("Vancouver", L1), [AFRICAN_DESCENT_SOCIETY, ALL_BODIES])
expect(from_city_only("Toronto", L1), [])

expect(from_city("Richmond", CHINESE_ORCHESTRA), True)
expect(from_city("Vancouver", CHINESE_ORCHESTRA), False)
expect(from_city("Abbotsford", BMX), True)
expect(from_city("Vancouver", BMX), False)

summary()
```
