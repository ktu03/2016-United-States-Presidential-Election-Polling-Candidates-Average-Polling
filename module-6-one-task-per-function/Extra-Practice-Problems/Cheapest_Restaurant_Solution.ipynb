```python
from cs103 import * 
from typing import NamedTuple
from typing import List
```

First, read the data definitions below.

```python
MenuItem = NamedTuple('MenuItem', [('name', str), 
                                   ('price', int)]) 
# interp. a menu item with its name and price in cents
M1 = MenuItem('coffee', 200)
M2 = MenuItem('muffin', 250)
M3 = MenuItem('sandwich', 675)
M4 = MenuItem('pizza', 250)
M5 = MenuItem('pop', 100)
M6 = MenuItem('beer', 650)
M7 = MenuItem('wings', 1200)
M8 = MenuItem('burger and fries', 1500)
M9 = MenuItem('Nachos', 1650)

# template based on compound
@typecheck
def fn_for_menu_item(mi: MenuItem) -> ...:
    return ... (mi.name,
                mi.price)


# List[MenuItem] 
# interp. a list of menu items
LOMI0 = []
LOMI1 = [M1, M2, M3]
LOMI2 = [M4, M5]
LOMI3 = [M6, M7, M8, M9]

# template based on arbitrary-sized and the reference rule
@typecheck
def fn_for_lomi(lomi: List[MenuItem]) -> ...:
    # description of the acc
    acc = ... # type: ...
    for mi in lomi:
        ...(acc,fn_for_menu_item(mi))
    return ...(acc)


Restaurant = NamedTuple('Restuarant', [('name', str), ('menu', List[MenuItem])])
# interp. A restaurant with name and list of menu items
# ASSUME: a restuarant has at least one menu item
R1 = Restaurant('Arbutus Coffee', LOMI1)
R2 = Restaurant('Uncle Fatihs Pizza', LOMI2)
R3 = Restaurant('Darbys Public House', LOMI3)

# template based on compound and the reference rule
@typecheck
def fn_for_restaurant(r: Restaurant) -> ...:
    return ... (r.name,
                fn_for_lomi(r.menu))
```

## Problem: 

Design a function that takes two restuarants and produces the name of the restaurant that has the lowest priced item on its menu. Be sure to follow all of the helper rules. 

```python
@typecheck
def has_cheapest_item(r1: Restaurant, r2: Restaurant) -> str:
    """
    return the name of the restaurant with the cheapest menu item
    """
    # return ''  #stub
    # template from Restaurant
    
    if price_of_cheapest_item(r1) < price_of_cheapest_item(r2):
        return r1.name
    else:
        return r2.name

@typecheck
def price_of_cheapest_item(r: Restaurant) -> int:
    """
    return the price of the cheapest item on the restaurant's menu
    """
    # return 0   #stub
    
    # template from Restuarant
    return cheapest_item(r.menu)

@typecheck
def cheapest_item(lomi: List[MenuItem]) -> int:
    """
    return the price of the cheapest item in the list
    ASSUME: the list has at least one element (as per the data definition)
    """
    #return 0  #stub
    # Template based on List[MenuItem]
    
    # cheapest is the cheapest menu item seen so far
    # since the list is assumed to have at least one element, we "peek"
    # at the first element to start
    cheapest = lomi[0]   # type: MenuItem
    for mi in lomi:
        if is_cheaper(mi, cheapest):
            cheapest = mi
    return cheapest.price  # No helper function needed to JUST get the price field

@typecheck
def is_cheaper(mi1: MenuItem, mi2: MenuItem) -> bool:
    """
    return True if mi1 is cheaper than mi2 and False otherwise
    """
    #return True  #stub
    # Template based on two MenuItems
    return mi1.price < mi2.price

# Begin testing
start_testing()

# Examples and tests for has_cheapest_item
expect(has_cheapest_item(R1, R2), 'Uncle Fatihs Pizza')
expect(has_cheapest_item(R1, R3), 'Arbutus Coffee')
expect(has_cheapest_item(R2, R3), 'Uncle Fatihs Pizza')

# Examples and tests for price_of_cheapest_item
expect(price_of_cheapest_item(R1), 200)
expect(price_of_cheapest_item(R2), 100)
expect(price_of_cheapest_item(R3), 650)

# Examples and tests for cheapest_item
expect(cheapest_item(LOMI1), 200)
expect(cheapest_item(LOMI2), 100)
expect(cheapest_item(LOMI3), 650)

# Examples and tests for is_cheaper
expect(is_cheaper(M1, M2), True)
expect(is_cheaper(M1, M1), False)
expect(is_cheaper(M2, M1), False)

# show testing summary
summary()
```
