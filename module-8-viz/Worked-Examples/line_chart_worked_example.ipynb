# Line Chart

This is an example of how to plot a line chart. In this example, we use some imaginary ecological data (fish caught over time). There are many ways to customize a pyplot line chart. If you have trouble, please ask for help. Try posting on our discussion board!

```python
# Note that here we're importing matplotlib.pyplot as pyplot. That means we need to use the name 
# pyplot whenever we want to use the pyplot library. In other examples, we've imported it as plt 
# and that would mean we need to use the name plt whenever we want to use the pyplot library.
import matplotlib.pyplot as pyplot
from cs103 import *
from typing import List

@typecheck
def plot_fish_over_time(fish_caught: List[int], hours: List[int]) -> None:
    """
    display a plot of the number of fish caught over time, where fish_caught is the number of fish caught 
    each hour (so, 0 fish caught after 0 hours, 2 fish caught after 1 hour, 7 fish caught after 2 hours, etc.)
    and hours is a list of all the hours for which there is data about the number of fish caught
    
    Assumes the lengths of fish_caught and hours are the same. Assumes that hours is provided in 
    increasing order. (Note: whatever hours is in, the lines connect in that order. So, decreasing order
    would work fine as well, but if you provided, say, [2, 3, 1], you'd see a line that went right and
    then back to the left!)
    """
    #return None #stub
    # Template based on visualization

    # set the x-axis label, y-axis label, and plot title
    pyplot.xlabel('hours')
    pyplot.ylabel('fish caught')
    pyplot.title('Fish caught over time')

    # range for the axes
    # [x-min, x-max, y-min, y-max]
    pyplot.axis([0,4,0,12])

    # plot our data 
    line = pyplot.plot(hours, fish_caught)

    # set some properties for the line (color to red, line width to 2, and marker to a small circle)
    pyplot.setp(line, color='r', linewidth=2.0, marker="o")

    # show the plot
    pyplot.show()
    
    return None
    
start_testing()

# Should produce an empty graph with x-axis "hours", y-axis "fish caught",
# and title "Fish caught over time":
expect(plot_fish_over_time([], []), None)

# See the sketch in the markdown cell below for what we expect this to look like.
expect(plot_fish_over_time([2, 8, 8], [1, 3, 4]), None)

summary()
```

```python
# Note that here we're importing matplotlib.pyplot as pyplot. That means we need to use the name 
# pyplot whenever we want to use the pyplot library. In other examples, we've imported it as plt 
# and that would mean we need to use the name plt whenever we want to use the pyplot library.
import matplotlib.pyplot as pyplot
from cs103 import *
from typing import List

@typecheck
def plot_fish_over_time_two_people(fish_caught_p1: List[int], 
                                   fish_caught_p2: List[int], 
                                   hours: List[int]) -> None:
    """
    display a plot of the number of fish caught over time, where fish_caught_p1 is the number of fish caught 
    each hour by person 1 (so, 0 fish caught after 0 hours, 2 fish caught after 1 hour, 7 fish caught after 2 hours, etc.)
    fish_caught_p2 is the number of fish caught each hour by person 2 and hours is a list of all the hours for 
    which there is data about the number of fish caught
    
    Assumes the lengths of fish_caught_p1, fish_caught_p2, and hours are all the same.
    """
    #return None #stub
    # Template based on visualization

    # set the x-axis label, y-axis label, and plot title
    pyplot.xlabel('hours')
    pyplot.ylabel('fish caught')
    pyplot.title('Fish caught over time')

    # range for the axes
    # [x-min, x-max, y-min, y-max]
    pyplot.axis([0,4,0,12])

    # plot our data 
    p1 = pyplot.plot(hours, fish_caught_p1)
    p2 = pyplot.plot(hours, fish_caught_p2)
    
    # set some properties for the lines, p1 and p2 (color to red, line width to 2, 
    # marker to a small circle, and a label for each)
    pyplot.setp(p1, color='r', linewidth=2.0, marker="o", label="Fisherperson 1")
    pyplot.setp(p2, color='b', linewidth=2.0, marker="o", label="Fisherperson 2")
    
    # We want to show the legend because this plot contains two lines. 
    # The first argument gives the x and y coordinates of where to place the legend box
    # while the second argument sets the legend to appear in the top right corner.
    # Want to learn more about different options you can use to configure the legend?
    # Google for "matplotlib.pyplot.legend documentation" and you can find the official
    # MatPlotLib site that will give you more detail. Another way you can do this is by
    # calling help(pyplot.legend) in a separate cell in Jupyter. It will give you all
    # the documentation right away!
    pyplot.legend(bbox_to_anchor=(1.05, 1), loc=2)

    # show the plot
    pyplot.show()
    
    return None
    
start_testing()

# Should produce an empty graph with x-axis "hours", y-axis "fish caught",
# title "Fish caught over time", and a legend showing fishers 1 and 2:
expect(plot_fish_over_time_two_people([], [], []), None)


# Should produce a graph with labels as above; content sketched in the 
# markdown cell below.
expect(plot_fish_over_time_two_people([2, 8, 8], [8, 2, 3], [1, 3, 4]), None)


summary()
```
