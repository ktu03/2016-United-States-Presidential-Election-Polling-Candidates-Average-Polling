# Scatterplot

Here, we're showing the relationship between (made up) people's ages and their salaries. 

On the plot, we use the size of each marker to indicate how many people have that age and salary. A size of 20 indicates one person, a size of 40 indicates two, and so on. We designed a function to convert the counts of people to areas of points.

```python
from cs103 import *
from typing import List


# List[int]
# interp. a list of numbers
L0 = []
L1 = [0, 3, 1, 6, 1]

# template based on arbitrary-sized data
@typecheck
def fn_for_loi(loi: List[int]) -> ...:
    # Describe what your accumulator means, e.g.: the ... so far
    acc = ...     # type: ...
    
    for i in loi:
        acc = ...(acc, i)
    
    return ...(acc)
```

```python
# We use this function to convert the raw data we're working with to useful numbers for our plot.
@typecheck
def convert_counts_to_areas(counts: List[int]) -> List[int]:
    """
    given a list of the counts of people of particular ages with particular salaries,
    return the appropriate corresponding areas of points to display in a chart
    
    Assumes all elements of counts are >= 0.
    """
    #return counts  #stub
    #template from List[int]
    
    # The list of converted counts so far
    acc = []     # type: List[int]
    
    for count in counts:
        # You could have converted the count directly here,
        # but it's also reasonable to use a helper function.
        acc.append(convert_count_to_area(count))
    
    return acc

@typecheck
def convert_count_to_area(count: int) -> int:
    """
    given a count of people of particular age with particular salary,
    return the appropriate area of a point to display in a chart
    
    Assumes count >= 0.
    """
    #return count  #stub
    #template based on atomic distinct
    return count*200
    

start_testing()

expect(convert_counts_to_areas([]), [])
expect(convert_counts_to_areas([3,9,0,2,0]), [60,180,0,40,0])

expect(convert_count_to_area(0), 0)
expect(convert_count_to_area(3), 60)

summary()
```

```python
# Note that here we're importing matplotlib.pyplot as plt. That means we need to use the name 
# plt whenever we want to use the pyplot library. In other examples, we've imported it as pyplot 
# and that would mean we need to use the name pyplot whenever we want to use the pyplot library.
import matplotlib.pyplot as plt

@typecheck
def show_scatterplot(ages: List[int], salaries: List[int], counts: List[int]) -> None:
    """
    display a scatterplot of salaries vs. ages. salaries are given in 1000s
    
    Assumes that the lengths of ages, salaries, and counts are all equal
    """
    #return None #stub
    # Template based on visualization
    
    areas = convert_counts_to_areas(counts)

    # set the labels for the axes
    plt.xlabel('Age')
    plt.ylabel('Salary (in 1000s)')
    plt.title('Salaries by age')

    # range for the axes
    # [x-min, x-max, y-min, y-max]
    plt.axis([0,65,0,105])

    # create the scatterplot, with markers that are red (c='r') and triangular (marker='^')
    plt.scatter(ages,salaries,marker='^', c='r', s=areas)

    # show the plot
    plt.show()
    
    return None

start_testing()

# Should display an empty graph with x-axis "Age", y-axis "Salary (in 1000s)", and title "Salaries by age".
expect(show_scatterplot([],  # ages
                        [],  # salaries
                        []), # counts
       None)     

# Should display a graph labelled much the same above. Contents sketched in the markdown cell below.
expect(show_scatterplot([10, 40],  # ages
                        [20, 60],  # salaries
                        [10,  2]), # counts
       None)     

summary()
```
