# Pie Chart

Here is an example pie chart showing the results of a (made up) survey that asked people about their favourite food. In this imaginary survery, 50% of people said their favourite food was burgers, 25% said pizza, 20% said soup, and 5% said noodles.

```python
import matplotlib.pyplot as plt
from cs103 import *
from typing import List

@typecheck
def show_pie_chart(labels: List[str], fractions: List[int], explode: List[float]) -> None:
    """
    show a pie chart with the given labels and corresponding fractions.
    
    "explode" (set out from the circle) each piece of the pie as specified in explode.
    
    fractions must each be in the range [0,100] and must sum to 100
    (and so must have at least one element).
    
    labels and fractions must have the same length.
    
    returns None
    """
    #return None  #stub
    # Template based on visualization

    # specify the percentages, then whether to explode each piece, then the labels, then the formatting
    # for the percentages on the chart, and finally whether you want a shadow or not. If you
    # remove autopct, the percentage will not show on each piece of the pie
    plt.pie(fractions, explode=explode, labels=labels, autopct='%.0f%%', shadow=False)

    plt.show()
    
    return None
    
start_testing()

# Should display a pie chart (roughly a circle) that is all just one thing (100%) and labelled "Universe".
expect(show_pie_chart(['Universe'], [100], [0]), None)

# Should display a pie chart with about 3/4 labelled "Dark Energy", about 1/4 labelled "Dark Matter", and a sliver sticking
# out labellled '"Normal" Matter'
expect(show_pie_chart(['Dark Matter', 'Dark Energy', '"Normal" Matter'], [23, 73, 4], [0, 0, 0.2]), None)

summary()
```
