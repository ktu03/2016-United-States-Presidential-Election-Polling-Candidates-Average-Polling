## Problem: 


Imagine you are studying the population cycles of predator and prey. One example of predator-prey interaction is between the lynx and snowshoe hare populations in Canada. To visualize this relationship, you need to plot the populations of these two species on the same graph using the following data (note that population numbers are given in thousands).

Be sure to give your graph a title, a legend, axes labels, and contrasting colours for the two plotted lines.

Data source: 
http://www-rohan.sdsu.edu/~jmahaffy/courses/f09/math636/lectures/lotka/qualde2.html#predatorprey

```python
# Given data 
time_in_years = [1900, 1901, 1902, 1903, 1904, 1905, 1906, 1907, 1908, 1909, 
                1910, 1911, 1912, 1913, 1914, 1915, 1916, 1917, 1918, 1919, 1920]

hare_population = [30, 47.2, 70.2, 77.4, 36.3, 20.6, 18.1, 21.4, 22, 25.4, 27.1, 
                   40.3, 57, 76.6, 52.3, 19.5, 11.2, 7.6, 14.6, 16.2, 24.7]

lynx_population = [4, 6.1, 9.8, 35.2, 59.4, 41.7, 19, 13, 8.3, 9.1, 7.4, 
                   8, 12.3, 19.5, 45.7, 51.1, 29.7, 15.8, 9.7, 10.1, 8.6]
```

```python
from matplotlib import pyplot
from cs103 import *
from typing import List
import matplotlib.ticker as ticker # needed to make the x-axis integers only

@typecheck
def show_pred_prey_chart(years: List[int], prey: List[float], pred: List[float]) -> None:
    """
    show the predator/prey cycle via a line chart (given years in which we have population data,
    and prey and predator populations in those years (in thousands)).
    
    Assumes the lengths of years, prey, and pred are equal and years is in increasing order.
    """
    # return None #stub
    # Template from visualization
    pyplot.xlabel('Time in Years')
    pyplot.ylabel('Population (in thousands)')
    pyplot.title('Populations of Lynx and Snowshoe Hare 1900-1920')
    hare = pyplot.plot(years, prey)
    lynx = pyplot.plot(years, pred)
    pyplot.setp(hare, color='g', linewidth=2.0, marker="o", label="Snowshoe Hare")
    pyplot.setp(lynx, color='r', linewidth=2.0, marker="x", label="Lynx")
    
    # we want to show the legend because this plot contains two bar charts.  
    # The arguments will place the legend to the right of the plot 
    pyplot.legend(loc='upper right') 
    
    # code to make sure the years are shown as integers
    pyplot.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
    
    pyplot.show()

    return None

start_testing()

# Should produce an empty graph with x-axis "Time in Years",
# y-axis "Population (in thousands)", title "Populations of Lynx and Snowshoe Hare 1900-1920",
# and a legend showing which is Snowshoe Hare and which is Lynx.
expect(show_pred_prey_chart([], [], []), None)

# Should produce a graph with the same labels as above but looking like the
# sketch in the markdown cell below:
expect(show_pred_prey_chart([2000,2010], [0, 10], [40, 30]), None)

summary()
```
